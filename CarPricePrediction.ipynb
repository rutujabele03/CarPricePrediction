```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ["name","selling_price","year","fuel","seller_type","transmission","owner","max_power"])
x
```

```python
y = df["selling_price"]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x.select_dtypes(include=np.number),y,test_size=0.20,random_state=99)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling = StandardScaler()
```

```python
x_train = scaling.fit_transform(x_train)
x_train
```

```python
x_test = scaling.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=10)
```

```python
rfr.fit(x_train,y_train)
```

```python
y_hat = rfr.predict(x_test)
y_hat
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
acc = r2_score(y_test,y_hat)
print(acc)
```
